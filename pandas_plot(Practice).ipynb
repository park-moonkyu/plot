```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re

# 데이터 로딩
df=pd.read_csv("./example.csv")

# 데이터 확인
df.head(10)

long=len(df['price'])

#정규화
for i in range(long):
    slt=re.findall("[0-9]", df['price'][i])
    df['price'][i]=int(''.join(slt))
    
df=df.head(10)
```

# 그래프 그리기

x, y 축 희망하는 컬럼명을 적어준다.<br>
그후 kind로 다양한 그래프를 구현할수있다.<br>
kind에는 line, hist,box,kde,are,pie,scatter 다 가능하다.<br>


```python
#line 그래프 그리기
df.plot('date','price', kind='line')
```

```python
#box 그래프 그리기
df.plot('date','price', kind='box')
```

```python
#scatter  그래프 그리기

ax=df.plot('date','price', kind='scatter')
```

```python
#hist  그래프 그리기
# grid로 눈금도 가능
ax2=df.plot('date','price', kind='hist',grid=True)
```
