```python
# watermark image

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


penguins = sns.load_dataset("penguins")

#아일랜드에는 총 3개가 있다. 
penguins['island'].unique()

# bill_length_mm, bill_depth_mm 두개의 축으로 3개의 island의 산점도를 알아보자
penguins.head()
```

# figure and axes
matplotlib으로 도화지 figure를 깔고 축공간 axes를 만든다.
1X2 축공간을 구성한다.


```python
fig, axes= plt.subplots(figsize=(4,4))
fig.tight_layout()

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",hue="species",data=penguins)
```
